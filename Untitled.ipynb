```python
def load_profiles_CMASS(usecols,home,suite,suiteset,sims,snap,prof):
    y=[]
    for s in np.arange(len(sims)):
        f=home+suite+'/'+suite+'_'+sims[s]+'_'+snap+'_w.txt'
        x,yi=np.loadtxt(f,usecols=usecols,unpack=True)
        yi=cgs_units(prof,yi)
        y.append(np.log10(yi))

    y=np.array(y)
    return x,y
```

```python
usecols_w_dict={'rho_mean':(0,1),'pth_mean':(0,5)}

#CMASS emulator scripts
def load_profiles_CMASS_mine(usecols,home,suite,suiteset,sims,snap,prof):
    y=[]
    for s in np.arange(len(sims)):
        f=home+suite+'/'+suiteset+'/'+suite+'_'+sims[s]+'_'+snap+'_w_CMASS.txt'
        x,yi=np.loadtxt(f,usecols=usecols,unpack=True)
        yi=fs.cgs_units(prof,yi)
        y.append(np.log10(yi))

    y=np.array(y)
    return x,y


def build_emulator_CMASS_mine(home,suite,suiteset,vary_str,prof,func_str):
    z=fs.choose_redshift(suite)
    z=z[-1]
    vary,sims=fs.choose_vary(vary_str)
    sims.remove('1P_27')
    vary = vary[vary!=1]
    print(sims, vary)
    snap='24'
    samples=vary
    nsamp=len(samples)

    usecols=usecols_w_dict[prof]
    x,y=load_profiles_CMASS_mine(usecols,home,suite,suiteset,sims,snap,prof)
    y=np.transpose(y)

    vary=vary[0:10]
    print(vary.shape, y.shape)
    
    emulator=ostrich.emulate.PcaEmulator.create_from_data(
        samples,
        y,
        ostrich.interpolate.RbfInterpolator,
        interpolator_kwargs={'function':func_str},
        num_components=12)
    return samples,x,y,emulator
```

```python
home='/home/jovyan/home/emu_CAMELS/emulator_profiles/' #point to your profiles
suiteset='1P'
suite='IllustrisTNG'
vary_str='ASN1'
prof='pth_mean' #rho_mean,rho_med,pth_mean,pth_med
func_str='linear' #this is the Rbf interpolation function

mass=fs.mass
mass_str=fs.mass_str
snap=fs.snap
z=fs.choose_redshift(suite)
vary,sims=fs.choose_vary(vary_str)
samples=fs.cartesian_prod(vary,z,mass) 
nsamp=samples.shape[0]

samples,x,y,emulator=build_emulator_CMASS_mine(home,suite,suiteset,vary_str,prof,func_str)
```
